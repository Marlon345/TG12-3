## 3. API-Endpunkt (JSON senden + empfangen)

````python
from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/api/data', methods=['POST'])
def get_data():
    incoming = request.get_json()
    response = {"received": incoming}
    return jsonify(response)

app.run(debug=True)
````

### JavaScript

```javascript
document.getElementById("myForm").addEventListener("submit", async (e) => {
    e.preventDefault();

    const text = document.getElementById("textInput").value;

    const res = await fetch("/api/data", {
        method: "POST",
        headers: {"Content-Type": "application/json"},
        body: JSON.stringify({text})
    });

    const json = await res.json();
    document.getElementById("output").innerText = JSON.stringify(json);
});
```

````python
class Katze:
    def __init__(self, name, alter):
        self.name = name
        self.alter = alter

    def miauen(self):
        return f"{self.name} sagt Miau!"
````

## 10. Beispiel-API mit Objektliste

````python
from flask import Flask, request, jsonify

app = Flask(__name__)

class Katze:
    def __init__(self, name):
        self.name = name

katzen = []

@app.route('/add', methods=['POST'])
def add():
    data = request.get_json()
    katzen.append(Katze(data['name']))
    return {"status": "ok"}

@app.route('/all')
def all_katzen():
    return jsonify([k.name for k in katzen])

app.run(debug=True)
````
